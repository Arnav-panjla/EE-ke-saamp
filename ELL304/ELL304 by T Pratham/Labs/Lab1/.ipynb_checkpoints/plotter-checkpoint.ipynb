```python
fig, ax = plt.subplots()
ax.plot(V_sg, np.sqrt(I_d))
ax.set_xlabel('$V_{SG}$')
ax.set_ylabel('$I_D$')
ax.set_title('pMOS Characteristics in saturation')
plt.show()

result = sp.linregress(V_sg, np.sqrt(I_d))
print(result.intercept)
print(result.slope)
print(result.intercept/result.slope)
fig, ax = plt.subplots()
plt.axline((result.intercept, 0), slope=result.slope)
ax.plot(V_sg, np.sqrt(I_d))
plt.xlim(-1, 9)
plt.ylim(-1, 6)
plt.show()
```
