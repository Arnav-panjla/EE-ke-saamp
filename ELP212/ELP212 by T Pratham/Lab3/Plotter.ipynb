```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
C = np.array([39, 47, 270, 1000, 10000, 22000, 100000, 220000])
Vs = np.array([696, 704, 664, 704, 816, 976, 1060, 328])

fig, ax = plt.subplots()
ax.plot(C, Vs)
ax.set_xlabel('Capacitance (pF)')
ax.set_ylabel('Output Voltage (mV)')
plt.title('Part 1: Capacitance vs Output voltage')
fig.savefig('plot1.png', dpi=300)
```
